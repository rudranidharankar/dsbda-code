```python
import pandas as pd

# Load the dataset
df = pd.read_csv('iris.csv')

# Filter the dataset for the specified species
selected_species = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
filtered_df = df[df['species'].isin(selected_species)]

# Group by species and compute descriptive statistics
species_stats = filtered_df.groupby('species').describe()

# Display the relevant statistics
print("Basic Statistical Details for Selected Species:")
print(species_stats)
```
